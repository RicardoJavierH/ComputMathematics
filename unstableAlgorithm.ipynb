```python
import numpy as np
np.set_printoptions(precision=4,suppress=True)
N = 5000
s = np.zeros(N)
s[0] = 1/2
print('    i         s(i)     (1/2)^i          s(i)-(1/2)^i  ')
for i in range(1,N) :
    s[i] = s[i-1]/2
    print(i, "  " , s[i], "  " ,np.power(1/2,i+1), "  ", s[i] - np.power(1/2,i+1) )
```
